```python
#Import Libraries
import sys
import re
import nltk

print('Python: \t', sys.version)
print('Author: \t Lawrance Koh (@LK)')
```

```python
def summarize_article(filename):
    file = open(filename, 'rb')
    text = file.read().decode('utf8')
    file.close()

    #remove binary encoding by force typing to string
    text = str(text) #force type into string

    # Removing Square Brackets and Extra Spaces
    text = re.sub(r'\[[0-9]*\]', ' ', text)
    text = re.sub(r'\s+', ' ', text)

    # Removing special characters and digits
    ftext = re.sub('[^a-zA-Z]', ' ', text )
    ftext = re.sub(r'\s+', ' ', ftext)
    
    sentence_list = nltk.sent_tokenize(text)
    stopwords = nltk.corpus.stopwords.words('english')

    word_frequencies = {}
    for word in nltk.word_tokenize(ftext):
        if word not in stopwords:
            if word not in word_frequencies.keys():
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1

    maximum_frequency = max(word_frequencies.values())

    for word in word_frequencies.keys():
        word_frequencies[word] = (word_frequencies[word]/maximum_frequency)
        
    #calculate sentence scores
    sentence_scores = {}
    for sent in sentence_list:
        for word in nltk.word_tokenize(sent.lower()):
            if word in word_frequencies.keys():
                #print(word, ": ", word_frequencies[word])
                if len(sent.split(' ')) < 30:
                    if sent not in sentence_scores.keys():
                        sentence_scores[sent] = word_frequencies[word]
                    else:
                        sentence_scores[sent] += word_frequencies[word]

    import heapq
    summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

    summary = ' '.join(summary_sentences)
    return(summary)
```

```python
filenames = ['Chapter A - Artificial Intelligence.txt', 
             'Chapter B - BigTech or the ANT-ification of Finance.txt',
             'Bank of the Future.txt',
             '1- Innovating in an Evolving World.txt',
             '2- All-Solid-State Batteries.txt',
             '3- Anti Aging Medicines.txt',
             '4 - Autonomous Vehicle Networks.txt',
             '5 - Big Data and Healthcare.txt',
             '6 - Dynamic Spectrum Access.txt',
             '7 - eSports.txt',
             '8 - 5G Technology.txt',
             '9 - Floating Offshore Wind Farms.txt',
             '10 - Real Estate Market Disruptors.txt',
             '11 - Smart Voice-Activated Assistants.txt',
             '12 - Disruptive Innovatios from the Past.txt'
            ]

for filename in filenames:
    print('Summarize:', filename, '\n')
    print(summarize_article(filename))
    print('*'*50)
```
